## 3. Feature Extraction and Selection

### Recursive feature elimination (RFE) with random forest

```python
# data_new.to_csv(r'C:Data.csv', index=False)
```

```python
from sklearn import preprocessing
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
```

```python
X = data_new.apply(lambda col: preprocessing.LabelEncoder().fit_transform(col))
X=X.drop(['Company','price'],axis=1)
y=data_new['price']
```

```python
# Create the RFE object and rank each pixel
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
clf_rf_3 = RandomForestRegressor()      
rfe = RFE(estimator=clf_rf_3, n_features_to_select=16, step=1)
rfe = rfe.fit(X, y)
print('Chosen best 16 feature by rfe:',X.columns[rfe.support_])
```

**After checking with different numbers we got a decent accuracy with Top 16 Features**

```python
features=list(X.columns[rfe.support_])
features
```

SO these are the top Best 16 Features that we will use in modeling

```python
data_new.head()
```

```python
# Lets create a new Dataframe with Our best Independent Features and Target Varibale
df = data_new[features]
df['price'] = data_new['price']
df
```

```python
df.info()
```

```python
# Lets categories the columns in categorical and numerical columns

cat_cols = df.select_dtypes(include='object').columns
num_cols = df.select_dtypes(exclude='object').columns
```

```python
cat_cols
```

### Dummy Variables

```python
df = pd.get_dummies(df, drop_first = True)
df.head()
```

```python
# Split our data into X and Y

X = df.drop('price',axis=1)
y = df.price
```

```python
### Train-Test Split

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X, y,
                                                 test_size=0.33, 
                                                 random_state = 45)
```

```python
x_train.head()
```

## Modeling and Evaluation

### Linear Regerssion

```python
from sklearn.linear_model import LinearRegression

lreg = LinearRegression()
lreg.fit(x_train,y_train)
y_train_pred = lreg.predict(x_train)
y_test_pred = lreg.predict(x_test)
lreg.score(x_test,y_test)
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

dt_regressor = DecisionTreeRegressor(random_state=0)
dt_regressor.fit(x_train,y_train)
y_train_pred = dt_regressor.predict(x_train)
y_test_pred = dt_regressor.predict(x_test)
dt_regressor.score(x_test,y_test)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

Rf = RandomForestRegressor(n_estimators = 15,
                              criterion = 'mse',
                              random_state = 20,
                              n_jobs = -1)
Rf.fit(x_train,y_train)
Rf_train_pred = Rf.predict(x_train)
Rf_test_pred = Rf.predict(x_test)


Rf.score(x_test, y_test)
```

### Conclusion
**Random forest performing best (with accuracy ~ 0.90)**
